##### **Bivariate Analysis**

```
power_by_turbine = train_data.groupby('cloud_level')['windmill_generated_power(kW/h)'].agg(['mean']).reset_index()
power_by_turbine.plot(kind='bar',x= 'cloud_level',y= 'mean',figsize = (9,4))
```

```
power_by_turbine = train_data.groupby('turbine_status')['windmill_generated_power(kW/h)'].agg(['mean']).reset_index()
power_by_turbine.plot(kind='bar',x= 'turbine_status',y= 'mean',figsize = (9,4))
```

```
obj_cols = train_data.select_dtypes('object').columns.values
numerical_cols = train_data[train_data.columns[~train_data.columns.isin(obj_cols)].values]

numerical_cols.head()
numerical_cols.shape
```

```
#Plot Correlation matrix to check for multi-collinearity
corrMatrix=numerical_cols.corr()
plt.figure(figsize=(20,8))

sns.heatmap(corrMatrix,annot=True)
plt.show()
```
